# Importar librerias

```python
import requests
import pandas as pd
```

# Leer URL y obtener solo los resultados deseados

```python
base_url = 'https://rss.itunes.apple.com/api/v1/us/apple-music/new-releases/all/100/explicit.json'
r = requests.get(base_url)
inf_dict = r.json()
music_list = inf_dict['feed']['results']
```

# Cargar los datos en un dataframe

```python
df = pd.json_normalize(music_list)
df
```

## Generar estadisticas por artista

```python
artist_dict = {}
for element in df.groupby('artistName'):
    artist_dict[element[0]] = element[1].shape[0]
print (artist_dict)
```

## Generar estadisticas por tipo

```python
kind_dict = {}
for element in df.groupby('kind'):
    kind_dict[element[0]] = element[1].shape[0]
print (kind_dict)
```

## Generar estadisticas por fecha de lanzamiento

```python
release_dict = {}
for element in df.groupby('releaseDate'):
    release_dict[element[0]] = element[1].shape[0]
print (release_dict)
```

## Canciones analizadas

```python
total = df.shape[0]
print (total)
```

## Estadisticas por genero

```python
genres_dict = {}
for song in music_list:
    for g in song['genres']:
        if g['name'] in genres_dict:
            genres_dict[g['name']] = genres_dict[g['name']] + 1
        else:
            genres_dict[g['name']] = 1
print (genres_dict)
```

## Diccionario de resumen

```python
resume_dict = {
    'total_songs': total,
    'release': release_dict,
    'genres': genres_dict,
    'artist': artist_dict,
}
print (resume_dict)
```
